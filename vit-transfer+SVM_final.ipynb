```python
import tensorflow as tf
from tensorflow.keras import layers
import numpy as np
import os
from sklearn.svm import SVC
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score, f1_score
import matplotlib.pyplot as plt
import seaborn as sns
import pickle
from tqdm import tqdm
from transformers import TFViTModel

# =================================================================
# 1. CU HNH THAM S
# =================================================================
IMG_HEIGHT = 224
IMG_WIDTH = 224
INPUT_SHAPE = (IMG_HEIGHT, IMG_WIDTH, 3)
NUM_CLASSES = 7

BATCH_SIZE = 32
EPOCHS = 50
LEARNING_RATE = 0.001

PRETRAINED_MODEL = "google/vit-base-patch16-224"

vit_weights_path = '/kaggle/working/vit_transfer_model.weights.h5'
plot_history_path = '/kaggle/working/training_curves.png'

print("="*70)
print("VIT TRANSFER LEARNING (PRE-TRAINED) + SVM CLASSIFIER")
print("="*70)
print(f"Using pre-trained model: {PRETRAINED_MODEL}")
print("="*70)

# =================================================================
# 2. NH NGHA LP BC (KHC PHC LI KERASTENSOR V CHANNEL ORDERING)
# =================================================================
class ViTFeatureExtractorLayer(tf.keras.layers.Layer):
    """Gi TFViTModel vo lp Keras Subclass  x l KerasTensor v gi hm."""
    def __init__(self, model_name, **kwargs):
        super().__init__(**kwargs)
        self.vit_model = TFViTModel.from_pretrained(
            model_name,
            output_hidden_states=True, 
            output_attentions=False
        )
    def call(self, inputs):
        outputs = self.vit_model(pixel_values=inputs, training=False)
        return outputs.pooler_output

# =================================================================
# 3. NH NGHA DATA PIPELINE & AUGMENTATION
# =================================================================
def load_and_preprocess_image(filepath, label):
    image = tf.io.read_file(filepath)
    image = tf.image.decode_image(image, channels=3, expand_animations=False)
    image.set_shape([None, None, 3])
    image = tf.image.resize(image, [IMG_HEIGHT, IMG_WIDTH])
    image = tf.cast(image, tf.float32) / 255.0
    return image, label

data_augmentation = tf.keras.Sequential([
    layers.RandomFlip("horizontal"),
    layers.RandomRotation(0.1),
    layers.RandomZoom(0.1),
    layers.RandomContrast(0.1),
])

# =================================================================
# 4. XY DNG MODEL VI PRE-TRAINED VIT
# =================================================================
def build_vit_transfer_learning(num_classes=NUM_CLASSES, trainable_layers='all'):
    
    vit_feature_layer = ViTFeatureExtractorLayer(model_name='google/vit-base-patch16-224')
    vit_base = vit_feature_layer.vit_model

    if trainable_layers == 'head':
        vit_base.trainable = False
    elif trainable_layers == 'all':
        vit_base.trainable = True
    elif isinstance(trainable_layers, int):
        vit_base.trainable = True
        for layer in vit_base.layers[:-trainable_layers]:
            layer.trainable = False
    
    inputs = layers.Input(shape=INPUT_SHAPE, name='pixel_values')
    
    x = layers.Normalization(
        mean=[0.5, 0.5, 0.5],
        variance=[0.25, 0.25, 0.25]
    )(inputs)

    # KHC PHC LI CHANNEL ORDERING: (N, H, W, C) -> (N, C, H, W)
    x = layers.Permute((3, 1, 2))(x)
    
    features = vit_feature_layer(x)
    
    features = layers.Dense(256, activation="gelu", name="feature_dense_1")(features)
    features = layers.Dropout(0.5, name="feature_dropout_1")(features)
    features = layers.Dense(128, activation="gelu", name="feature_dense_2")(features)
    features = layers.Dropout(0.4, name="feature_dropout_2")(features)
    
    outputs = layers.Dense(num_classes, activation="softmax", name="predictions")(features)
    
    full_model = tf.keras.Model(inputs=inputs, outputs=outputs, name="ViT_Transfer_Full")
    
    feature_extractor = tf.keras.Model(
        inputs=inputs,
        outputs=full_model.get_layer("feature_dense_2").output,
        name="ViT_Transfer_FeatureExtractor"
    )
    
    return full_model, feature_extractor, vit_base

# =================================================================
# 5. LOAD D LIU
# =================================================================
def load_image_paths_labels(data_dir):
    class_names = sorted(os.listdir(data_dir))
    paths = []
    labels = []
    for i, cls in enumerate(class_names):
        cls_dir = os.path.join(data_dir, cls)
        if not os.path.isdir(cls_dir):
            continue
        for fname in os.listdir(cls_dir):
            if fname.lower().endswith(('.png', '.jpg', '.jpeg')):
                paths.append(os.path.join(cls_dir, fname))
                labels.append(i)
    return paths, labels, class_names

train_dir = '/kaggle/input/flowerdatasets/flowers/train'
val_dir = '/kaggle/input/flowerdatasets/flowers/val'
test_dir = '/kaggle/input/flowerdatasets/flowers/test'

train_paths, train_labels, class_names = load_image_paths_labels(train_dir)
val_paths, val_labels, _ = load_image_paths_labels(val_dir)
test_paths, test_labels, _ = load_image_paths_labels(test_dir)

print(f"\nClasses: {class_names}")
print(f"Train samples: {len(train_paths)}")
print(f"Validation samples: {len(val_paths)}")
print(f"Test samples: {len(test_paths)}")

train_labels_one_hot = tf.keras.utils.to_categorical(train_labels, num_classes=NUM_CLASSES)
val_labels_one_hot = tf.keras.utils.to_categorical(val_labels, num_classes=NUM_CLASSES)
test_labels_one_hot = tf.keras.utils.to_categorical(test_labels, num_classes=NUM_CLASSES)

# =================================================================
# 6. TO DATA PIPELINE
# =================================================================
train_ds = tf.data.Dataset.from_tensor_slices((train_paths, train_labels_one_hot))
train_ds = train_ds.shuffle(len(train_paths)) \
    .map(load_and_preprocess_image, num_parallel_calls=tf.data.AUTOTUNE) \
    .map(lambda x, y: (data_augmentation(x, training=True), y), num_parallel_calls=tf.data.AUTOTUNE) \
    .batch(BATCH_SIZE) \
    .prefetch(tf.data.AUTOTUNE)

val_ds = tf.data.Dataset.from_tensor_slices((val_paths, val_labels_one_hot))
val_ds = val_ds.map(load_and_preprocess_image, num_parallel_calls=tf.data.AUTOTUNE) \
    .batch(BATCH_SIZE) \
    .prefetch(tf.data.AUTOTUNE)

test_ds = tf.data.Dataset.from_tensor_slices((test_paths, test_labels_one_hot))
test_ds = test_ds.map(load_and_preprocess_image, num_parallel_calls=tf.data.AUTOTUNE) \
    .batch(BATCH_SIZE) \
    .prefetch(tf.data.AUTOTUNE)

# =================================================================
# 7. TRAINING VIT (HEAD-ONLY)
# =================================================================
print("\n" + "="*70)
print("BC 1: TRAINING HEAD CA VIT TRANSFER LEARNING MODEL (HEAD-ONLY)")
print("="*70)

# CU HNH HEAD-ONLY
RETRAIN_VIT = True
FINE_TUNE_VIT = False
TRAINABLE_LAYERS = 'head'

full_model, feature_extractor, vit_base = build_vit_transfer_learning(
    num_classes=NUM_CLASSES,
    trainable_layers=TRAINABLE_LAYERS
)

# t li trng thi FREEZE
vit_base.trainable = False 
print(" Backbone ViT  b ng bng (Frozen). Ch hun luyn Classifier Head.")
print(f"  - Total params: {full_model.count_params():,}")
print(f"  - Trainable params: {sum([tf.size(w).numpy() for w in full_model.trainable_weights]):,}")

# --- HM V BIU  MI ---
def plot_training_history(history, filename):
    plt.figure(figsize=(15, 5))

    # Loss plot
    plt.subplot(1, 2, 1)
    plt.plot(history.history['loss'], label='Train Loss')
    plt.plot(history.history['val_loss'], label='Validation Loss')
    plt.title('Loss Curve')
    plt.xlabel('Epoch')
    plt.ylabel('Loss')
    plt.legend()
    plt.grid(True)

    # Accuracy plot
    plt.subplot(1, 2, 2)
    plt.plot(history.history['accuracy'], label='Train Accuracy')
    plt.plot(history.history['val_accuracy'], label='Validation Accuracy')
    plt.title('Accuracy Curve')
    plt.xlabel('Epoch')
    plt.ylabel('Accuracy')
    plt.legend()
    plt.grid(True)

    plt.tight_layout()
    plt.savefig(filename, dpi=300, bbox_inches='tight')
    plt.show()

if RETRAIN_VIT:
    print("\n Training Classifier Head (Frozen Backbone)...")
    
    full_model.compile(
        optimizer=tf.keras.optimizers.AdamW(learning_rate=LEARNING_RATE),
        loss='categorical_crossentropy',
        metrics=['accuracy']
    )
    
    callbacks = [
        tf.keras.callbacks.EarlyStopping(
            monitor='val_accuracy',
            patience=10,
            restore_best_weights=True,
            verbose=1
        ),
        tf.keras.callbacks.ReduceLROnPlateau(
            monitor='val_loss',
            factor=0.2,
            patience=5,
            min_lr=1e-7,
            verbose=1
        ),
        tf.keras.callbacks.ModelCheckpoint(
            filepath=vit_weights_path,
            monitor='val_accuracy',
            save_best_only=True,
            save_weights_only=True,
            verbose=1
        )
    ]
    
    history = full_model.fit(
        train_ds,
        epochs=EPOCHS,
        validation_data=val_ds,
        callbacks=callbacks,
        verbose=1
    )
    
    print(" Training Head hon tt!")
    # V BIU  SAU KHI TRAIN
    plot_training_history(history, plot_history_path)

else:
    print("\n OPTION 3: Load weights c sn...")
    if os.path.exists(vit_weights_path):
        full_model.load_weights(vit_weights_path)
        print(" Load weights thnh cng!")
    else:
        print(" KHNG tm thy weights!")
        raise FileNotFoundError(f"Weights file not found: {vit_weights_path}")

print("\n nh gi ViT Transfer Learning trn test set:")
full_model.compile(
    optimizer=tf.keras.optimizers.AdamW(learning_rate=LEARNING_RATE),
    loss='categorical_crossentropy',
    metrics=['accuracy']
)
test_loss, test_acc = full_model.evaluate(test_ds, verbose=0)
print(f"ViT Transfer Learning Test Accuracy: {test_acc * 100:.2f}%")

# =================================================================
# 8. TRCH XUT FEATURES VI VIT
# =================================================================
print("\n" + "="*70)
print("BC 2: TRCH XUT FEATURES T VIT")
print("="*70)

def extract_features_vit(model, image_paths, batch_size=32):
    features_list = []
    
    def load_image_only(filepath):
        image = tf.io.read_file(filepath)
        image = tf.image.decode_image(image, channels=3, expand_animations=False)
        image.set_shape([None, None, 3])
        image = tf.image.resize(image, [IMG_HEIGHT, IMG_WIDTH])
        image = tf.cast(image, tf.float32) / 255.0
        return image
    
    ds = tf.data.Dataset.from_tensor_slices(image_paths)
    ds = ds.map(load_image_only, num_parallel_calls=tf.data.AUTOTUNE) \
        .batch(batch_size) \
        .prefetch(tf.data.AUTOTUNE)
    
    for batch in tqdm(ds, desc="Extracting features"):
        batch_features = model(batch, training=False).numpy()
        features_list.append(batch_features)
    
    features = np.concatenate(features_list, axis=0)
    return features

print("\nTrch xut features cho tp train...")
train_features = extract_features_vit(feature_extractor, train_paths)

print("\nTrch xut features cho tp validation...")
val_features = extract_features_vit(feature_extractor, val_paths)

print("\nTrch xut features cho tp test...")
test_features = extract_features_vit(feature_extractor, test_paths)

print(f"\n Kch thc features:")
print(f"  Train: {train_features.shape}")
print(f"  Val: {val_features.shape}")
print(f"  Test: {test_features.shape}")

# =================================================================
# 9. CHUN HA FEATURES
# =================================================================
print("\n" + "="*70)
print("BC 3: CHUN HA FEATURES")
print("="*70)

scaler = StandardScaler()
train_features_scaled = scaler.fit_transform(train_features)
val_features_scaled = scaler.transform(val_features)
test_features_scaled = scaler.transform(test_features)

print("  chun ha features vi StandardScaler")

# =================================================================
# 10. TRAINING SVM VI HYPERPARAMETERS C NH
# =================================================================
print("\n" + "="*70)
print("BC 4: TRAINING SVM CLASSIFIER")
print("="*70)

# Cu hnh hyperparameters c nh
SVM_CONFIG = {
    'C': 10,
    'gamma': 'scale',
    'kernel': 'rbf',
    'random_state': 42
}

print(f"\n SVM Hyperparameters (Fixed):")
print("-" * 70)
for key, value in SVM_CONFIG.items():
    print(f"  {key:<15} : {value}")
print("-" * 70)

print("\n Training SVM vi hyperparameters c nh...")
svm_classifier = SVC(**SVM_CONFIG)
svm_classifier.fit(train_features_scaled, train_labels)
print(" Training hon tt!")

# nh gi trn validation set
val_pred_svm = svm_classifier.predict(val_features_scaled)
val_acc_svm = accuracy_score(val_labels, val_pred_svm)
val_f1_svm = f1_score(val_labels, val_pred_svm, average='macro')

print(f"\n Validation Performance:")
print("-" * 70)
print(f"Validation Accuracy: {val_acc_svm * 100:.2f}%")
print(f"Validation F1-Score (Macro): {val_f1_svm * 100:.2f}%")
print("-" * 70)

# =================================================================
# 11. NH GI SVM
# =================================================================
print("\n" + "="*70)
print("BC 5: NH GI HIU SUT")
print("="*70)

test_pred_svm = svm_classifier.predict(test_features_scaled)
test_acc_svm = accuracy_score(test_labels, test_pred_svm)
test_f1_svm = f1_score(test_labels, test_pred_svm, average='macro')

print("\n" + "-"*70)
print("SO SNH KT QU")
print("-"*70)
print(f"{'Model':<35} {'Test Accuracy':<15}")
print("-"*70)
print(f"{'ViT Transfer + Softmax':<35} {test_acc*100:>6.2f}%")
print(f"{'ViT Transfer + SVM':<35} {test_acc_svm*100:>6.2f}%")
print("-"*70)

# =================================================================
# 12. CLASSIFICATION REPORT V CONFUSION MATRIX
# =================================================================
print("\n" + "="*70)
print("CHI TIT PHN TCH - VIT TRANSFER + SVM")
print("="*70)

print("\nClassification Report:")
print("-"*70)
print(classification_report(test_labels, test_pred_svm, target_names=class_names, digits=4))

def plot_confusion_matrix(y_true, y_pred, title, filename):
    cm = confusion_matrix(y_true, y_pred)
    
    plt.figure(figsize=(10, 8))
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues',
                xticklabels=class_names, yticklabels=class_names,
                cbar_kws={'label': 'Count'})
    plt.title(title, fontsize=16, fontweight='bold', pad=20)
    plt.xlabel('Predicted Label', fontsize=12, fontweight='bold')
    plt.ylabel('True Label', fontsize=12, fontweight='bold')
    plt.xticks(rotation=45, ha='right')
    plt.yticks(rotation=0)
    plt.tight_layout()
    plt.savefig(filename, dpi=300, bbox_inches='tight')
    plt.show()

plot_confusion_matrix(
    test_labels,
    test_pred_svm,
    'Confusion Matrix - ViT Transfer Learning + SVM',
    '/kaggle/working/confusion_matrix_vit_transfer_svm.png'
)

# =================================================================
# 13. HIN TH NH D ON NG V SAI
# =================================================================
print("\n" + "="*70)
print("PHN TCH D ON NG V SAI")
print("="*70)

def plot_predictions(image_paths, true_labels, pred_labels, class_names,
                     indices, title, filename, correct=True):
    fig, axes = plt.subplots(1, 3, figsize=(15, 5))
    fig.suptitle(title, fontsize=16, fontweight='bold', y=1.02)
    
    for idx, ax in zip(indices, axes):
        img = tf.keras.preprocessing.image.load_img(image_paths[idx])
        img_array = tf.keras.preprocessing.image.img_to_array(img)
        
        ax.imshow(img_array.astype('uint8'))
        ax.axis('off')
        
        true_class = class_names[true_labels[idx]]
        pred_class = class_names[pred_labels[idx]]
        
        color = 'green' if correct else 'red'
        symbol = '' if correct else ''
        
        ax.set_title(
            f'{symbol} Prediction: {pred_class}\nTrue: {true_class}',
            fontsize=11,
            fontweight='bold',
            color=color,
            pad=10
        )
    
    plt.tight_layout()
    plt.savefig(filename, dpi=300, bbox_inches='tight')
    plt.show()

correct_indices = np.where(test_pred_svm == test_labels)[0]
incorrect_indices = np.where(test_pred_svm != test_labels)[0]

print(f"\n Thng k:")
print(f"  - S nh d on NG: {len(correct_indices)}/{len(test_labels)} ({len(correct_indices)/len(test_labels)*100:.2f}%)")
print(f"  - S nh d on SAI: {len(incorrect_indices)}/{len(test_labels)} ({len(incorrect_indices)/len(test_labels)*100:.2f}%)")

if len(correct_indices) >= 3:
    correct_samples = np.random.choice(correct_indices, 3, replace=False)
    
    print("\n Hin th 3 nh d on NG:")
    plot_predictions(
        test_paths, test_labels, test_pred_svm, class_names,
        correct_samples,
        ' D ON NG - SVM PREDICTIONS',
        '/kaggle/working/correct_predictions.png',
        correct=True
    )
else:
    print("\n Khng  3 nh d on ng  hin th")

if len(incorrect_indices) >= 3:
    incorrect_samples = np.random.choice(incorrect_indices, 3, replace=False)
    
    print("\n Hin th 3 nh d on SAI:")
    plot_predictions(
        test_paths, test_labels, test_pred_svm, class_names,
        incorrect_samples,
        ' D ON SAI - SVM PREDICTIONS',
        '/kaggle/working/incorrect_predictions.png',
        correct=False
    )
else:
    print("\n Model hon ho! Khng c nh d on sai")


# =================================================================
# 14. LU M HNH - SAVEDMODEL FORMAT (FOLDER)
# =================================================================
print("\n" + "="*70)
print("LU M HNH - SAVEDMODEL FORMAT")
print("="*70)

# QUAN TRNG: Compile model trc khi lu
full_model.compile(
    optimizer=tf.keras.optimizers.AdamW(learning_rate=LEARNING_RATE),
    loss='categorical_crossentropy',
    metrics=['accuracy']
)

# Lu TON B model vi SavedModel format (FOLDER, khng phi .keras file)
print(" Saving complete model...")
try:
    full_model.save('/kaggle/working/vit_flower_model')
    print("  lu: vit_flower_model/ (SavedModel format)")
except Exception as e:
    print(f" Li khi lu: {e}")
    print("\n Th phng n d phng: Export inference-only model...")
    
    # To inference model (khng c training components)
    @tf.function(input_signature=[tf.TensorSpec(shape=[None, IMG_HEIGHT, IMG_WIDTH, 3], dtype=tf.float32)])
    def serving_fn(images):
        return full_model(images, training=False)
    
    # Save vi concrete function
    tf.saved_model.save(
        full_model,
        '/kaggle/working/vit_flower_model',
        signatures={'serving_default': serving_fn}
    )
    print("  lu: vit_flower_model/ (Inference-only)")

# Lu metadata
model_metadata = {
    'class_names': class_names,
    'img_size': (IMG_HEIGHT, IMG_WIDTH),
    'test_accuracy': test_acc,
    'pretrained_model': PRETRAINED_MODEL,
    'num_classes': NUM_CLASSES
}

with open('/kaggle/working/model_metadata.pkl', 'wb') as f:
    pickle.dump(model_metadata, f)
print("  lu: model_metadata.pkl")

# Check folder size
from pathlib import Path
folder_size = sum(f.stat().st_size for f in Path('/kaggle/working/vit_flower_model').rglob('*')) / (1024*1024)
print(f"\n Model info:")
print(f"  - Folder size: {folder_size:.1f} MB")
print(f"  - Test Accuracy: {test_acc * 100:.2f}%")

if folder_size < 100:
    print("\n WARNING: Folder size qu nh! ViT weights c th cha c lu!")
else:
    print(" Folder size hp l - ViT weights  c lu!")

# =================================================================
# 15. TEST LOAD V D ON
# =================================================================
print("\n" + "="*70)
print("DEMO: LOAD V D ON")
print("="*70)

# Load metadata
with open('/kaggle/working/model_metadata.pkl', 'rb') as f:
    metadata = pickle.load(f)

print(f" Loaded metadata")
print(f"  - Classes: {metadata['class_names']}")
print(f"  - Test Accuracy: {metadata['test_accuracy']*100:.2f}%")

# Load model
print("\n Loading model...")
is_savedmodel_format = False
try:
    # Th load vi custom_objects
    loaded_model = tf.keras.models.load_model(
        '/kaggle/working/vit_flower_model',
        custom_objects={'ViTFeatureExtractorLayer': ViTFeatureExtractorLayer}
    )
    print(" Model loaded successfully!")
except Exception as e:
    print(f" Khng load c vi keras API: {e}")
    print(" Th load vi SavedModel API...")
    
    # Load vi SavedModel API
    loaded_model = tf.saved_model.load('/kaggle/working/vit_flower_model')
    is_savedmodel_format = True
    print(" Model loaded vi SavedModel API!")
    print(" Note: Ch c th dng inference, khng train c")
    
    # Kim tra signatures
    print(f" Available signatures: {list(loaded_model.signatures.keys())}")

# =================================================================
# 16. TEST D ON V SO SNH
# =================================================================
print("\n" + "="*70)
print("TEST D ON")
print("="*70)

def predict_image(image_path, model, class_names, is_savedmodel=False):
    """D on class cho 1 nh"""
    image = tf.io.read_file(image_path)
    image = tf.image.decode_image(image, channels=3, expand_animations=False)
    image.set_shape([None, None, 3])
    image = tf.image.resize(image, [IMG_HEIGHT, IMG_WIDTH])
    image = tf.cast(image, tf.float32) / 255.0
    image = tf.expand_dims(image, 0)
    
    if is_savedmodel:
        # Dng SavedModel signature
        infer = model.signatures['serving_default']
        predictions = infer(image)
        # Output c th l dict, cn extract tensor
        if isinstance(predictions, dict):
            predictions = list(predictions.values())[0].numpy()
        else:
            predictions = predictions.numpy()
    else:
        # Keras API
        predictions = model.predict(image, verbose=0)
    
    pred_class = np.argmax(predictions[0])
    confidence = predictions[0][pred_class]
    
    return pred_class, confidence

# Test trn 10 nh ngu nhin
test_sample_size = min(10, len(test_paths))
sample_indices = np.random.choice(len(test_paths), test_sample_size, replace=False)

print(f"\n So snh d on gia ORIGINAL model v LOADED model:")
print("-"*80)
print(f"{'True Class':<15} {'Original':<15} {'Conf':<8} {'Loaded':<15} {'Conf':<8} {'Match':<10}")
print("-"*80)

correct_original = 0
correct_loaded = 0
predictions_match = 0

for idx in sample_indices:
    path = test_paths[idx]
    true_class = test_labels[idx]
    
    # Predict vi original model
    pred_orig, conf_orig = predict_image(path, full_model, class_names, is_savedmodel=False)
    
    # Predict vi loaded model
    pred_load, conf_load = predict_image(path, loaded_model, metadata['class_names'], is_savedmodel=is_savedmodel_format)
    
    # Check accuracy
    if pred_orig == true_class:
        correct_original += 1
    if pred_load == true_class:
        correct_loaded += 1
    
    # Check if predictions match
    match = "" if pred_orig == pred_load else ""
    if pred_orig == pred_load:
        predictions_match += 1
    
    print(f"{class_names[true_class]:<15} {class_names[pred_orig]:<15} {conf_orig*100:>5.1f}% {' '*2} "
          f"{metadata['class_names'][pred_load]:<15} {conf_load*100:>5.1f}% {' '*2} {match}")

print("-"*80)
print(f"Original Model Accuracy: {correct_original}/{test_sample_size} ({correct_original/test_sample_size*100:.1f}%)")
print(f"Loaded Model Accuracy:   {correct_loaded}/{test_sample_size} ({correct_loaded/test_sample_size*100:.1f}%)")
print(f"Predictions Match:       {predictions_match}/{test_sample_size} ({predictions_match/test_sample_size*100:.1f}%)")

if predictions_match == test_sample_size:
    print("\n PERFECT: Original v Loaded model cho kt qu GING HT NHAU!")
elif correct_loaded >= test_sample_size * 0.95:
    print(f"\n GOOD: Loaded model hot ng tt ({correct_loaded/test_sample_size*100:.1f}% accuracy)")
else:
    print(f"\n ERROR: Loaded model c vn  nghim trng!")

# =================================================================
# 17. VERIFY TRN TON B TEST SET
# =================================================================
print("\n" + "="*70)
print("VERIFY TRN TON B TEST SET")
print("="*70)

print("\n Evaluating original model...")
test_loss_orig, test_acc_orig = full_model.evaluate(test_ds, verbose=0)

print(" Evaluating loaded model...")
if is_savedmodel_format:
    print(" SavedModel format - d on tng batch...")
    correct = 0
    total = 0
    infer = loaded_model.signatures['serving_default']
    for images, labels in test_ds:
        preds = infer(images)
        # Extract predictions from dict or tensor
        if isinstance(preds, dict):
            preds = list(preds.values())[0].numpy()
        else:
            preds = preds.numpy()
        pred_classes = np.argmax(preds, axis=1)
        true_classes = np.argmax(labels.numpy(), axis=1)
        correct += np.sum(pred_classes == true_classes)
        total += len(true_classes)
    test_acc_load = correct / total
    test_loss_load = 0.0
else:
    test_loss_load, test_acc_load = loaded_model.evaluate(test_ds, verbose=0)

print("\n RESULTS:")
print("-"*70)
print(f"Original Model: Loss = {test_loss_orig:.4f}, Accuracy = {test_acc_orig*100:.2f}%")
print(f"Loaded Model:   Loss = {test_loss_load:.4f}, Accuracy = {test_acc_load*100:.2f}%")
print(f"Difference:     Accuracy = {abs(test_acc_orig - test_acc_load)*100:.4f}%")
print("-"*70)

if abs(test_acc_orig - test_acc_load) < 0.01:
    print(" PERFECT: Model c load chnh xc!")
elif test_acc_load > 0.95:
    print(" GOOD: Model hot ng tt!")
else:
    print(" ERROR: C vn  khi load model!")

```
